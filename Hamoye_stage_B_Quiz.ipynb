```python
# import all necessary modules
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# read csv file into pandas
data = pd.read_csv("energydata_complete.csv")

#drop unwanted columns
data = data.drop(["date", "lights"], axis=1)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

# normalise data
normalised_df = pd.DataFrame(scaler.fit_transform(data), columns=data.columns)

# set target variale (Y) and predictors(X)
y = normalised_df["Appliances"]
x = normalised_df.drop("Appliances", axis=1)
```

```python
# From the dataset, fit a linear model on the relationship between the 
# temperature in the living room in Celsius (x = T2) and the temperature outside the building (y = T6).
# What is the R^2 value in two D.P?
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# set target variale (Y) and predictors(X)
# y = np.array(normalised_df["T6"]).reshape((-1, 1))
# x = np.array(normalised_df["T2"]).reshape((-1, 1))

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)

# instantiate the a linear_model object
linear_model = LinearRegression()

#fit the model to the training dataset
linear_model.fit(x_train, y_train)

#obtain predictions
y_pred = linear_model.predict(x_test)
```

```python
# calculate the r-squared
from sklearn.metrics import r2_score
r2_score = r2_score(y_test, y_pred)
round(r2_score, 2) 
```

```python
# calculate the MAE
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_test, y_pred)
round(mae, 2)
```

```python
#calculate the RSS
import numpy as np
rss = np.sum(np.square(y_test - y_pred))
round(rss, 2)
```

```python
#calculate the RMS
from sklearn.metrics import  mean_squared_error
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
round(rmse, 3)
```

```python
# ridge Regression
# Train a ridge regression model with an alpha value of 0.4. 
# Is there any change to the root mean squared error (RMSE) when evaluated on the test set?

from sklearn.linear_model import Ridge
ridge_reg = Ridge(alpha=0.4)
ridge_reg.fit(x_train, y_train)

#calculate the RMS
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
round(rmse, 3)
```

```python
#Lasso Regression
# Train a lasso regression model with an alpha value of 0.001 and obtain the new feature weights with it. 
# How many of the features have non-zero feature weights?

from sklearn.linear_model import Lasso
lasso_reg = Lasso(alpha=0.001)
lasso_reg.fit(x_train, y_train)

#What is the new RMSE with the Lasso Regression (in 3 decimal places)?  
#calculate the RMS
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
round(rmse, 3)
```

```python
#comparing the effects of regularisation
def get_weights_df(model, feat, col_name):
    #this function returns the weight of every feature
    weights = pd.Series(model.coef_, feat.columns).sort_values()
    weights_df = pd.DataFrame(weights).reset_index()
    weights_df.columns = ['Features', col_name]
    weights_df[col_name].round(3)
    return weights_df

# Obtain the feature weights from your linear model above
linear_model_weights = get_weights_df(linear_model, x_train, 'Linear_Model_Weight')
ridge_weights_df = get_weights_df(ridge_reg, x_train, 'Ridge_Weight')
lasso_weights_df = get_weights_df(lasso_reg, x_train, 'Lasso_weight')

final_weights = pd.merge(linear_model_weights, ridge_weights_df, on='Features')
final_weights = pd.merge(final_weights, lasso_weights_df, on='Features')
final_weights
```
